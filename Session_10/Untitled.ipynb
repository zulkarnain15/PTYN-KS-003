```python
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns

sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
from scipy.stats import uniform
```

```python
n = 1000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
```

```python
ax = sns.distplot(data_uniform, bins=100, kde=True, color='red', hist_kws={'linewidth': 15, 'alpha':1})
ax.set(xlabel = 'Uniform Distribution ', ylabel='Frequency')
plt.show()
```

```python
from scipy.stats import norm
data_normal = norm.rvs(size=n, loc=0, scale=1)
norm_plot = sns.distplot(data_normal, bins=20, color='blue', kde=False)
```

```python
ax = sns.distplot(data_normal, bins=100, kde=True, color='red', hist_kws={'linewidth': 15, 'alpha':1})
ax.set(xlabel = 'Normal Distribution ', ylabel='Frequency')
plt.show()
```
