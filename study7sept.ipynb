```python
movieDB = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movieDB.head()
```

```python
movieDB.describe()
```

```python
movieDB.info()
```

```python
#filter film dengan duration <= 100, star_rating > 7, 
# urutkan dari rating tertinggi

movieDBfilter = movieDB[(movieDB['duration'] < 100) & (movieDB['star_rating'] > 7)]
```

```python
movieDBfilter = movieDBfilter.sort_values(by=['star_rating'])
movieDBfilter.head()
```

```python
movieDBfilter[movieDBfilter.title.str.contains('^s', flags=re.IGNORECASE)]
```

```python
groupedMovie = movieDB.groupby(['genre'])['title']
```

```python
for key,value in groupedMovie.groups.items():
    for valInx in value:
        print(movieDB[movieDB.genre == key])
```

```python
summary = movieDB.pivot_table(index=['genre'], 
                              values= ['duration', 'star_rating'], 
                              aggfunc={'duration':'mean', 
                                       'star_rating':'mean'})
```

```python
summary.plot.bar(rot = 0)
```
